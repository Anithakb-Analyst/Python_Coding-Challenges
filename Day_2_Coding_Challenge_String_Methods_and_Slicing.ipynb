# Feedback Analyzer


Content review team wants to analyse user feedback and comments

```python
Feedback = input("Please enter your feedback: ")
```

```python
Feedback_words = Feedback.split()
Words_count = len(Feedback_words)
print(Feedback_words)
print ('Words_count:', Words_count)
```

```python
data_count = Feedback.count('data')
print("Number of 'data' occurence:", data_count)
```

```python
print('First_5 Characters: ', Feedback[:5])
```

```python
print('Last_5 Characters: ', Feedback[-5:])
```

```python
print('Reverse of the sentence; ', Feedback[::-1])
```
